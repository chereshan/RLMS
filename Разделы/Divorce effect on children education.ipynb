```python
def merge_df_i_hh(year):
    df_i, df_h=RENAMED_IND_DFS[year][RENAMED_IND_DFS[year].under18==1],RENAMED_HH_DFS[year]
    merged_df=df_i.reset_index().merge(df_h, on=f'{waves_dict[year][1].lower()}id_h', how='left',suffixes=[None,'_hh']).set_index('index')
    return merged_df
```

```python
def family_structure(ind_column):
    fin_list=[ind_column]
    prefix=ind_column[0]
    num=int(str(ind_column[1:]).replace('idind',''))
    year=reverse_waves_dict[prefix.upper()][0]
    r2=max([int(i.replace('idind','').replace(waves_dict[year][1].lower(),'')) for i in RENAMED_HH_DFS[year].columns[RENAMED_HH_DFS[year].columns.str.contains('idind')]])
    
    for y in range(1,num):
        if num==1:
            continue
        else:
            if y<10:
                s=f'b{num}.9.{y}'
            else:
                if num<17:
                    s=f'b{num}9.{y}'
                else:
                    s=f'b{num}.9{y}'
            fin_list.append(s)
    for y in range(num+1,r2+1):
        if num<10:
            s=f'b{y}.9.{num}'
        else:
            if y<17:
                s=f'b{y}9.{num}'
            else:
                s=f'b{y}.9{num}'
        fin_list.append(s)
            

    return fin_list
```

```python
def parent_counter(tabl):
    return pd.Series({'Bio_parents':(tabl=='Сын/дочь').sum()+(tabl=='Отец/мать').sum(),
    'Foster_parents':(tabl=='Неродной сын/неродная дочь').sum()+(tabl=='Отчим/мачеха').sum()})
    
```

```python
def ch_ind_columns(year):
    df_i, df_h=RENAMED_IND_DFS[year][RENAMED_IND_DFS[year].under18==1],RENAMED_HH_DFS[year]
    idind_columns=[i for i in df_h.columns[df_h.columns.str.contains('idind')]]
    merged_df=df_i.reset_index().merge(df_h, on=f'{waves_dict[year][1].lower()}id_h', how='left').set_index('index')
    idind_table=merged_df.loc[:,idind_columns].isin(df_i.idind).idxmax(axis=1)
    return idind_table
```

```python
def parents_df(year):
    df_i, df_h=RENAMED_IND_DFS[year][RENAMED_IND_DFS[year].under18==1],RENAMED_HH_DFS[year]
    idind_columns=[i for i in df_h.columns[df_h.columns.str.contains('idind')]]
    merged_df=df_i.reset_index().merge(df_h, on=f'{waves_dict[year][1].lower()}id_h', how='left').set_index('index')
    idind_table=merged_df.loc[:,idind_columns].isin(df_i.idind).idxmax(axis=1)
    return pd.DataFrame({ind: parent_counter(merged_df.loc[ind,family_structure(idind_table[ind])]) for ind in idind_table.index}).T
```

What I did:
1. We transform individual dataframes id-columns of RLMS to int form beacuse of how Pyhton non-ideally mathc float=float. int=int is much better in python. 
2. Same for households dataframes.
3. We locate every child (<18) raw in individual dataframes
4. We merge them together with associated household dataframes so evey individual child raw has columns of their household information
5. Then I develop alghorithm that construct list of columns sufficient to check all relations in family of a given child
6. After that we count number of parents (Biological and/or a foster) of a raw. 
7. Repeat same operation for every raw of every dataframe of every year of RLMS-survey
8. We constuct 2 new columns "number of biological parens" and "number of foster parents"
9. To speed things up a little we create global variables of merged children's individual and household data

```python
import statsmodels.formula.api as smf

# OLS results including interaction terms:
reg = smf.ols(formula='delta_educ ~ C(h5)+C(status)+C(nonfull_fm)+C(under7)+C(over18)+nfm+income_per_fm+I(income_per_fm**2)+apart_per_fm', data=Full_10years)
results = reg.fit()
results.summary()
```

### Interpretation

* ``Intercept `` - years of getted of educations after 10 years of a child all things being equal (sex=female, status=city, family=full, older then 7 years, younger then 18 years)
* ``C(h5)[T.МУЖСКОЙ]`` - sex of a child (effect of male sex on years of education after 10 years)
* ``C(status)``- city, countryside, town
* ``C(nonfull_fm)`` - single biologial parent household 
* ``C(under7)`` - younger then 7 years
* ``C(over18)`` - older then 18 years 
* ``nfm`` - number of family members in household
* ``income_per_fm`` - incomer per family member of households
* ``apart_per_fm`` - apartment area per family member of household, town
